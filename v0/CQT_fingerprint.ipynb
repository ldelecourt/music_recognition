# Audio signal analysis with CQT

### Data extraction from .wav file

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import librosa.core as librosa
import librosa.display as librosa_display

data, rate = librosa.load("test.wav", sr=44100, mono=True) # mono=True transforme l'audio en mono
```

### Constant Q Transform

```python
# with bins_per_octave=36 and the sample rate = 44100
cqt = librosa.cqt(data, sr=rate, bins_per_octave=36)
log_cqt = librosa.amplitude_to_db(np.abs(cqt))
```

### CQT spectrogram

```python
# plot cqt
plt.figure(2, figsize=(12,6))
plt.subplot(211)
librosa_display.specshow(librosa.amplitude_to_db(np.abs(cqt)), sr=rate, x_axis='time', y_axis='hz', cmap='plasma')
plt.colorbar(format='%+2.0f dB')
plt.tight_layout()
plt.title("CQT")
# plot log_cqt
plt.subplot(212)
librosa_display.specshow(log_cqt, sr=rate, x_axis='time', y_axis='hz', cmap='plasma')
plt.colorbar(format='%+2.0f dB')
plt.tight_layout()
plt.title("log CQT")
plt.show()
```
