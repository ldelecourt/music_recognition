```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import librosa.core 
import librosa.display

data, sr = librosa.load("audio.mp3", sr=44100, mono=True) # mono=True transforme l'audio en mono
```

```python
onset_frames = librosa.onset.onset_detect(data, sr=sr, wait=1, pre_avg=1, post_avg=1, pre_max=1, post_max=1)
print(onset_frames) # frame numbers of estimated onsets

onset_times = librosa.frames_to_time(onset_frames)
print(onset_times)
```

```python
S = librosa.stft(data)
logS = librosa.amplitude_to_db(abs(S))

librosa.display.specshow(librosa.amplitude_to_db(logS, ref=np.max), y_axis='log', x_axis='time')
plt.show()
```
