## Audio signal analysis with FFT

### Data extraction from .wav file

```python
%matplotlib inline
import numpy as np
import scipy.io.wavfile
import matplotlib.pyplot as plt
from numpy import fft as fft

rate, data = scipy.io.wavfile.read("test.wav")
```

### Determine if the audio is stereo or mono

```python
print(np.shape(data))
```

So it is stereo, cause data as 2 columns.

### Extract the two channels

```python
channel1 = data[:,0] #left
channel2 = data[:,1] #right
```

### Plot amplitude over time

```python
#create a time variable in seconds
time = np.arange(0, float(data.shape[0]), 1) / rate 

# Plot
plt.figure()
plt.subplot(211)
plt.plot(time, channel1, linewidth=0.02, alpha=0.9, color='blue')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.subplot(212)
plt.plot(time, channel2, linewidth=0.02, alpha=0.9, color='blue')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()
```

### Fast Fourier Transform exemple

```python
fourier = fft.fft(channel1)
print(fourier)
```

### Plot FFT spectrogram (the transformation is done into the plt.specgram function)

```python
plt.figure(2, figsize=(12,6))
plt.subplot(211)
spectrum1, freqs, bins, im = plt.specgram(channel1, Fs=rate, NFFT=1024, cmap=plt.get_cmap('plasma'))
cbar=plt.colorbar(im)
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
cbar.set_label('Intensity dB')
plt.subplot(212)
spectrum2, freqs, bins, im = plt.specgram(channel2, Fs=rate, NFFT=1024, cmap=plt.get_cmap('plasma'))
cbar=plt.colorbar(im)
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
cbar.set_label('Intensity (dB)')
plt.show()
```

### Finding peaks in spectrum1

```python
from scipy.ndimage.filters import maximum_filter
from scipy.ndimage.morphology import (generate_binary_structure, iterate_structure, binary_erosion)

struct = generate_binary_structure(2, 1)
neighborhood = iterate_structure(struct, 20)
```

```python
# find local maxima using our filter shape
local_max = maximum_filter(spectrum1, footprint=neighborhood) == spectrum1
```

```python
print(np.shape(spectrum1))
print(np.shape(local_max))
```

```python
# Erosion
background = (spectrum1 == 0)
eroded_background = binary_erosion(background, structure=neighborhood, border_value=1)
```

```python
# Boolean mask of arr2D with True at peaks (Fixed deprecated boolean operator by changing '-' to '^')
detected_peaks = local_max ^ eroded_background

# extract peaks
amps = spectrum1[detected_peaks]
j, i = np.where(detected_peaks)
```

```python
# filter peaks
amp_min = 10

amps = amps.flatten()
peaks = zip(i, j, amps)
peaks_filtered = filter(lambda x: x[2]>amp_min, peaks) # freq, time, amp
# get indices for frequency and time
frequency_idx = []
time_idx = []
for x in peaks_filtered:
    frequency_idx.append(x[1])
    time_idx.append(x[0])

peaks_data = zip(frequency_idx, time_idx)
```

### Extract peaks

```python
peaks = set(peaks_data)
```

```python
l_peaks = list(peaks)
```

```python
print(l_peaks)
print(np.shape(l_peaks))
```
