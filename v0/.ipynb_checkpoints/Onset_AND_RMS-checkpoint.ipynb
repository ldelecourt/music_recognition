```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import librosa.core 
import librosa.display

data, sr = librosa.load("audio.mp3", sr=44100, mono=True) # mono=True transforme l'audio en mono
```

# TEST ONSET

```python
plt.figure(figsize=(14, 5))
librosa.display.waveplot(data, sr=sr)
plt.show()
```

```python
hop_length = 256
onset_envelope = librosa.onset.onset_strength(data, sr=sr, hop_length=hop_length)
onset_envelope
```

```python
N = len(data)
T = N/float(sr)
t = np.linspace(0, T, len(onset_envelope))


plt.figure(figsize=(14, 5))
plt.plot(t, onset_envelope)
plt.xlabel('Time (sec)')
plt.xlim(xmin=0)
plt.ylim(0)
```

```python
onset_frames = librosa.util.peak_pick(onset_envelope, 7, 7, 7, 7, 0.5, 5)
```

```python
plt.figure(figsize=(14, 5))
plt.plot(t, onset_envelope)
plt.grid(False)
plt.vlines(t[onset_frames], 0, onset_envelope.max(), color='r', alpha=0.7)
plt.xlabel('Time (sec)')
plt.xlim(0, T)
plt.ylim(0)
```

# TEST RMS AVEC CQT

```python
fmin = librosa.midi_to_hz(36)
hop_length = 512
C = librosa.cqt(data, sr=sr, fmin=fmin, n_bins=72, hop_length=hop_length)

logC = librosa.amplitude_to_db(np.abs(C))
plt.figure(figsize=(15, 5))
librosa.display.specshow(logC, sr=sr, x_axis='time', y_axis='cqt_note', fmin=fmin, cmap='coolwarm')
```

```python
S, phase = librosa.magphase(logC)
rms = librosa.feature.rms(S=S)
```

```python
plt.figure()
plt.subplot(2, 1, 1)
plt.semilogy(rms.T, label='RMS Energy')
plt.xticks([])
plt.xlim([0, rms.shape[-1]])
plt.legend()
plt.subplot(2, 1, 2)
librosa.display.specshow(librosa.amplitude_to_db(S, ref=np.max), y_axis='log', x_axis='time')
plt.title('log Power spectrogram')
plt.tight_layout()
```
